```python
print('Results, grade_id vs:')
print('  * years_climbing:')
print('    - Coefficient: ', round(lm_results_years_climbing.params[1],2))
print('    - P-value: ', round(lm_results_years_climbing.pvalues[1], 3))
print('  * is_female:')
print('    - Coefficient: ', round(lm_results_is_female.params[1],2))
print('    - P-value: ', round(lm_results_is_female.pvalues[1], 3))
print('  * weight:')
print('    - Coefficient: ', round(lm_results_weight.params[1],2))
print('    - P-value: ', round(lm_results_weight.pvalues[1], 3))
print('  * height:')
print('    - Coefficient: ', round(lm_results_height.params[1],2))
print('    - P-value: ', round(lm_results_height.pvalues[1], 3))
```

All independent variables considered (years climbing, gender, weight, and height) have P-values << 0.05. This means that for any practical significance level, we reject the null hypothesis that these variables do not have an affect on the climbing grade.

However, what's not well established is how the grade_id value correlates to actual difficulty of a climb. In general, the higher the grade_id, the more difficult but those IDs account for four unique systems (US rope climbing, US bouldering, French rope climbing, and French bouldering). We need to consider this!

```python
raw_df_grade
```

Looking closer at it, the grade ID matches one-to-one with the French grading system. As such, using the grade ID as the dependent variable works perfectly!

```python
print('Results, grade_id vs:')
print('  * years_climbing:')
print('    - Coefficient: ', round(lm_results_years_climbing.params[1],2))
print('    - P-value: ', round(lm_results_years_climbing.pvalues[1], 3))
print('  * is_female:')
print('    - Coefficient: ', round(lm_results_is_female.params[1],2))
print('    - P-value: ', round(lm_results_is_female.pvalues[1], 3))
print('  * weight:')
print('    - Coefficient: ', round(lm_results_weight.params[1],2))
print('    - P-value: ', round(lm_results_weight.pvalues[1], 3))
print('  * height:')
print('    - Coefficient: ', round(lm_results_height.params[1],2))
print('    - P-value: ', round(lm_results_height.pvalues[1], 3))
```

```python
df.columns
```

```python
# ---- ROPE, YEARS, ALL ----
lm_results_rope_years_climbing_mf = smf.ols('grade_id ~ years_climbing', 
        data=df[(df['years_climbing']>=0) & 
                (df['years_climbing']<=50) &
                (df['is_bouldering']==0)
        ]).fit()
```

```python
# ---- ROPE, YEARS, MALE ----
lm_results_rope_years_climbing_m = smf.ols('grade_id ~ years_climbing', 
        data=df[(df['years_climbing']>=0) & 
                (df['years_climbing']<=50) &
                (df['is_bouldering']==0) &
                (df['is_female']==0)
        ]).fit()
```

```python
# ---- ROPE, YEARS, FEMALE ----
lm_results_rope_years_climbing_f = smf.ols('grade_id ~ years_climbing', 
        data=df[(df['years_climbing']>=0) & 
                (df['years_climbing']<=50) &
                (df['is_bouldering']==0) &
                (df['is_female']==1)
        ]).fit()
```

```python
lm_results_rope_years_climbing_mf.params
```

```python
lm_results_rope_years_climbing_mf.pvalues
```

```python
df_results = pd.DataFrame([{
    'Type': 'Rope',
    'Variable': 'Years climbing (0-50)',
    'Gender': 'MF',
    'Coefficient, LM': round(lm_results_rope_years_climbing_mf.params[1],3),
    'P-value, LM': round(lm_results_rope_years_climbing_mf.pvalues[1], 3)
}])

df_results = df_results[['Type','Variable','Gender','Coefficient, LM','P-value, LM']]

df_results
```

```python
df_results_temp = pd.DataFrame([{
        'Type': 'Rope',
        'Variable': 'Years climbing (0-50)',
        'Gender': 'M',
        'Coefficient, LM': round(lm_results_rope_years_climbing_m.params[1],3),
        'P-value, LM': round(lm_results_rope_years_climbing_m.pvalues[1], 3)
}])

df_results_temp = df_results_temp[['Type','Variable','Gender','Coefficient, LM','P-value, LM']]

df_results = pd.DataFrame.append(df_results, df_results_temp).reset_index(drop=True)

df_results
```

```python
df_results_temp = pd.DataFrame([{
        'Type': 'Rope',
        'Variable': 'Years climbing (0-50)',
        'Gender': 'F',
        'Coefficient, LM': round(lm_results_rope_years_climbing_f.params[1],3),
        'P-value, LM': round(lm_results_rope_years_climbing_f.pvalues[1], 3)
}])

df_results_temp = df_results_temp[['Type','Variable','Gender','Coefficient, LM','P-value, LM']]

df_results = pd.DataFrame.append(df_results, df_results_temp).reset_index(drop=True)

df_results
```
